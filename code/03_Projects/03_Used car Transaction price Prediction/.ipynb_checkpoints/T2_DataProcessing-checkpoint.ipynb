### 偏度和峰度
1. 偏度（Skewness）:描述数据分布形态的统计量，其描述的是某总体取值分布的对称性，简单来说就是数据的不对称程度。

偏度是三阶中心距计算出来的:
- Skewness = 0 ，分布形态与正态分布偏度相同。
- Skewness > 0 ，正偏差数值较大，为正偏或右偏。长尾巴拖在右边，数据右端有较多的极端值。
- Skewness < 0 ，负偏差数值较大，为负偏或左偏。长尾巴拖在左边，数据左端有较多的极端值。

数值的绝对值越大，表明数据分布越不对称，偏斜程度大。

计算公式： $$Skewness=E[((x-E(x))/(\sqrt{D(x)}))^3]$$
|Skewness| 越大，分布形态偏移程度越大。

2. 峰度（Kurtosis）:偏度是描述某变量所有取值分布形态陡缓程度的统计量，简单来说就是数据分布顶的尖锐程度。
峰度是四阶标准矩计算出来的:
- Kurtosis=0 与正态分布的陡缓程度相同。
- Kurtosis>0 比正态分布的高峰更加陡峭——尖顶峰
- Kurtosis<0 比正态分布的高峰来得平台——平顶峰

计算公式：$$Kurtosis=E[ ( (x-E(x))/ (\sqrt(D(x))) )^4 ]-3$$

**参考资料：**   
1. [偏度和峰度如何影响您的分布](https://support.minitab.com/zh-cn/minitab/18/help-and-how-to/statistics/basic-statistics/supporting-topics/data-concepts/how-skewness-and-kurtosis-affect-your-distribution/#skewness)
2. [数据的偏度和峰度](https://www.cnblogs.com/wyy1480/p/10474046.html)

```python
# 查看预测值price的分布
# print(train_data['price'].value_counts())

# price的整体分布概况
import scipy.stats as st
y = train_data['price']
plt.figure(1)
plt.title('约翰逊分布')
sns.distplot(y, kde=False, fit=st.johnsonsu)
plt.figure(2)
plt.title('正态分布')
sns.distplot(y, kde=False, fit=st.norm)
plt.figure(3)
plt.title('对数后正态分布')
sns.distplot(y, kde=False, fit=st.lognorm)
```

可以看出price不服从正态分布，属于偏态分布，因此在回归预测前需要对其进行转换，在对以上三图观察后，最佳拟合是无界约翰逊分布

```python
# 对price查看skewness(偏斜系数) and kurtosis(峰值)
sns.distplot(train_data['price'])
print('Skewness: %f' % (train_data['price'].skew()))
print("Kurtosis: %f" % (train_data['price'].kurt()))
```

```python
# 人为区分数值特征和类别特征
numeric_features = ['power', 'kilometer', 'v_0', 'v_1', 'v_2', 'v_3', 'v_4', 'v_5', 'v_6', 'v_7', 'v_8', 'v_9', 'v_10', 'v_11', 'v_12', 'v_13','v_14' ]
categorical_features = ['name', 'model', 'brand', 'bodyType', 'fuelType', 'gearbox', 'notRepairedDamage', 'regionCode']

# 类别特征的nunique分布
print('训练集中的类别特征分布如下：\n')
for cat_fea in categorical_features:
    print(cat_fea + '的特征分布如下：')
    print('{} 特征有{}个特征值'.format(cat_fea, train_data[cat_fea].nunique()))
    print(train_data[cat_fea].value_counts())
print('*-'*30)
    
# 数值特征分析
numeric_features.append('price')
print('训练集中数值特征有：\n')
print(numeric_features)
```

```python
# 相关性分析
price_numeric = train_data[numeric_features]
correlation_numeric = price_numeric.corr()
print(correlation_numeric['price'].sort_values(ascending=False), '\n')

figure, axis = plt.subplots(figsize=(7, 7))
plt.title('Correlation of Numeric Features with Price', y=1, size=16)
sns.heatmap(correlation_numeric, square=True, vmax=0.8)
```
