```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')

pd.set_option('float_format', lambda x: '%.6f' % x)
pd.set_option('Display.max_rows',None)#展示全部行
pd.set_option('Display.max_columns',None)
```

```python
df = pd.read_excel('../../dataset/AutiFraud/data_base.xlsx')

df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
# 统计Y值的分类数量并重新命名
label = df['Y'].value_counts()
label = label.reset_index()
label = label.rename(index={0:'label_1', 1:'label_2'}, columns={'index':'Y', 'Y': 'num'})
label['prct'] = label['num'] / len(df)
label
```

```python
# 缺失值处理
df['X11'] = df['X11'].fillna(df['X11'].mean())
df['X13'] = df['X13'].fillna(df['X13'].mean())
df['X14'] = df['X14'].fillna(df['X14'].mean())
df['X15'] = df['X15'].fillna(df['X15'].mean())
df['X16'] = df['X16'].fillna(df['X16'].mean())
df['X17'] = df['X17'].fillna(df['X17'].mean())
df['X18'] = df['X18'].fillna(df['X18'].mean())
df['X19'] = df['X19'].fillna(df['X19'].mean())
df['X20'] = df['X20'].fillna(df['X20'].mean())
df.isnull().sum()
```

```python
# 特征相关系数：pearson系数
corr = df.drop(columns=['ID', 'Y']).corr(method='pearson').round(4)
corr
```
