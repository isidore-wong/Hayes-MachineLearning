### 预测贷款用户是否会逾期

Task1: 对数据进行探索和分析，进行数据预处理

```python
import pandas as pd

df = pd.read_csv('../../dataset/LoanUsersData/data.csv', encoding='gbk')
df.head()
```

```python
df.info() # 共有4754条数据，90列，并且有些属性有缺失值，值为NA
```

```python
df.describe()
```

```python
# 数据预处理
df_cleaned = df.copy()
df_cleaned.drop(['Unnamed: 0','custid','trade_no','bank_card_no' ,'source','id_name'], inplace=True, axis=1)
print(len(df_cleaned.columns))
# df_cleaned.columns
```

```python
# 用其他城市填充未知NA，并将城市分布转换为数字
df_cleaned['reg_preference_for_trad'].fillna('其他城市', inplace=True)
df_cleaned['reg_preference_for_trad'].replace({'一线城市':1, '二线城市':2, '三线城市':3, '境外':4, '其他城市':5}, inplace=True)

# 查看非数值型特征的分布
print(df_cleaned['reg_preference_for_trad'].value_counts(dropna=False))
```

```python
# 缺失值处理
df_cleaned.isnull().sum()
```

```python
# 对缺失量较大的student_feature特征进行处理
print(df_cleaned['student_feature'].value_counts(dropna=False))
# 用0填充缺失值
df_cleaned['student_feature'].fillna(0, inplace=True)
print(df_cleaned['student_feature'].value_counts(dropna=False))
```

```python
# 其余特征的缺失值用众数填充
# 或采用的填充方式：数值型取中位数，离散型/日期取众数
for i in df_cleaned.columns:
    df_cleaned[i].fillna(df_cleaned[i].mode()[0], inplace=True)
df_cleaned.head()
```

```python
# 数据集切分
import numpy as np
from sklearn.model_selection import train_test_split

Y = df_cleaned['status']
X = df_cleaned.drop('status', axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2018)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
df_output = pd.concat([X, Y], axis=1)
df_output.head()
```
