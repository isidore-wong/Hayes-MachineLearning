```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from factor_analyzer import FactorAnalyzer
import warnings

warnings.filterwarnings('ignore')
```

```python
test_data = pd.read_csv('../../dataset/factor_analysis_dataset/test01.csv')

print(test_data.info())
print(test_data.shape)
test_data.head()
```

```python
# Pearson correlation 
test01_corr = test_data.corr()
print('\n特征间的相关系数：\n')
pd.DataFrame(test01_corr)
```

```python
# 矩阵R的特征值和特征向量
eig_value, eig_vector = np.linalg.eig(test01_corr)
eig = sorted(eig_value, reverse=True)    # 贡献率从大到小排序
print('\n特征值排序:\n', eig)

index_eig = sorted(range(len(eig_value)), key = lambda k : eig_value[k])

# 确定公因子个数，判断条件--前n个特征值的累积贡献率>=0.85, 确定公因子个数为16
for n in range(1, 20):
    if pd.Series(eig[:n]).sum() / pd.Series(eig).sum() >= 0.85:
        print('\n公因子个数：', n)
        break
```

```python
fa = FactorAnalyzer(n_factors=16, rotation=None)
# fa.analyze(test_data, 16, rotation=None)
fa.fit(test_data)
print('\n公因子方差为：\n')
pd.DataFrame(fa.get_communalities())
```
